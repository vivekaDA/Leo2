```python
#write a python program to extract and display all the header tags from en.wikipedia.org/wiki/Main_page
from bs4 import BeautifulSoup
import requests
url = requests.get('https://en.wikipedia.org/wiki/Main_Page')
soup = BeautifulSoup(url.text, 'html.parser')
story = soup.find_all(['h1', 'h2','h3'])
for i in story:
    file = i.text
    print(file)
```

```python
#write a program to extract and display all the image links from en.wikipedia.org/wiki/Peter_jeffrey_(RAAF_officer)
import requests
from bs4 import BeautifulSoup
url = requests.get('https://en.wikipedia.org/wiki/Peter_Jeffrey_(RAAF_officer)')
soup = BeautifulSoup(url.text, 'html.parser')
story = soup.find_all(['img'])
for i in story:
    file = i.get('src')
    print(file)
```

```python
#write a python program to get the number of followers of a given twitter account
import requests
from bs4 import BeautifulSoup
def get_twitter_followers(username):
    url = f'https://twitter.com/DonaldJ.Trump'
    response = requests.get(url)

    if response.status_code ==200:
        soup = BeautifulSoup(response.text, 'html.parser')
        followers_tag = soup.find('a', href=f'/DonaldJ.Trump/followers')
        if followers_tag:
            followers_count = followers_tag.find('span', class_='css-901oao css-16my406 r-1qd0xha r-ad9z0x r-bcqeeo r-qvutc0')
            if followers_count:
                return followers_count.text
            else:

                return "followers count not found"






```

```python
#write a program to display  the contains of different attributes like different attributes like different attributes like status_code,header,url,history,encoding,reason,cookies,elapsed,request and content of a specified resource
import requests
import json
url = requests.get('https://en.wikipedia.org/wiki/Main_Page')
print('Status code:',url.status_code)
print('Header:',url.headers)
print('URL:',url.url)
print('History:',url.history)
print('Encoding:',url.encoding)
print('Reason:',url.reason)
print('Cookies:',url.cookies)
print('Elapsed:',url.elapsed)
print('Request:',url.request)
print('Content:',url.content)
print('Text:',url.text)
print('JSON:',url.json())

```
